````python
# Inventario inicial:
inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "once"},
    {"producto": "pantalon", "precio": 39900, "stock": "23"},
]
````

````python
nuevo_producto = {
    "producto": "abrigo",
    "precio": 50000,
    "stock": "2"}

inventario.append(nuevo_producto)
````

````python
inventario
````

````python
## 1. Acceda el inventario (Recorrer la lista)
## 2. Acceda al diccionario que hay en cada posicion
## 3. Tome la llave stock y convierta su valor a entero
## 4. Reasigne el nuevo valor (entero) a la llave de stock para que se actualice

primer_producto = inventario[0]
primer_producto['stock'] = int(primer_producto.get("stock"))
primer_producto
````

````python
segundo_producto = inventario[1]
segundo_producto['stock'] = int(segundo_producto.get("stock"))
segundo_producto
inventario
````

````python
for prod in inventario:
  ## Validar que la transformacion sea correcta

  try:
    prod['stock'] = int(prod.get("stock"))
  except:
    print("Transformacion incorrecta - hay un valor que no puedo convertir a")

  #if(type(prod['stock']) == int):
  #  print("Transformacion correcta")
  #else:
  #  print("Transformacion incorrecta")
  prod['precio'] = prod['precio'] + 10000
  stock = prod['stock']
  producto = prod['producto']
  precio = prod['precio']
  print(f"Hay {stock} unidades del producto {producto}. Su precio por unidad es de {precio}")
````

````python
def cambiar_stock(diccionario):
  diccionario['stock'] = int(diccionario.get("stock"))
  return diccionario

nuevo_inventario = list(map(cambiar_stock,inventario))
nuevo_inventario
````

````python
def actualizar_precio(diccionario):
  diccionario['precio'] = diccionario.get("precio") + 10000
  return diccionario

nuevo_inventario = list(map(actualizar_precio,inventario))
nuevo_inventario
````

## Control de flujo

### Condiciones If, Elif y Else

````python
mi_numero = int(input("ingresa tu numero: "))

if mi_numero < 10: # La sentencia if evalua una condición, en caso de ser verdadera, se ejecuta el bloque de codigo que le sigue
  print("Tu número es menor que 10")
elif mi_numero > 60:

  if mi_numero % 2 == 0:
    print("Tu número es mayor que 60 y es par")
  else:
    print("Tu número es mayor que 60 y es impar")

elif mi_numero > 40:
  print("Tu número es mayor que 40")
elif mi_numero > 20:
  print("Tu número es mayor que 20")
else:
  print("Tu número está entre 10 y 20")
````

````python
## Condiciones con negación
mi_edad = int(input("ingresa tu edad: "))
if not mi_edad >= 18: # Evalua si mi edad NO es mayor o igual a 18
  print("Eres menor de edad")
else:
  print("Eres mayor de edad")
````

````python
## Condiciones multiples

mi_numero = int(input("ingresa tu numero: "))

if (mi_numero > 0 and mi_numero < 10) or mi_numero == 7: # Las condiciones se evaluan de izquierda a derecha
  print("Tu número es mayor que 0 y menor que 10")
elif mi_numero == 10 or mi_numero == 0:
  print("Tu número es 10 o 0")
else:
  print("Tu número es mayor a 10")
````

### Bucles For (Iteración basada en una secuencia de objetos)

````python
# Los bucles for generalmente se usan para recorrer una secuencia de objetos. El bloque de codigo que contiene el bucle se ejecutara por cada objeto que exista en la secuencia ----> ITERACION
estudiantes_notas = [("Cristian",5.0),("Lorena",4.0),("Lina",4.2),("Adrian",5.0),("Edwin",3.8),("Andrea",4.7)]

for item in estudiantes_notas: # Para cada tupla estudiante - nota en la lista de estudiantes_notas, realice las siguientes acciones
    print(f"{item[0]} obtuvo una nota de {item[1]} en el diplomado de análisis de datos")
    print("---------------------------------------------------------------------")
````

````python
suma_total = 0 #Declaro la variable suma_total que almacenara el acumulado de la suma

for numero in range(10,51,5):  # Ejemplo iterando sobre un rango de numeros. Se suma cada número y se imprime el resultado total.
  print(f"Imprimiendo el número:  {numero}")
  suma_total += numero

print("------------------------------------------")
print(f"La sumatoria total es: {suma_total}")
````

````python
suma_total = 0 #Declaro la variable suma_total que almacenara el acumulado de la suma

# Uso de las palabras claves break y continue

for numero in range(100):  # Ejemplo iterando sobre un rango de numeros del 0 al 10.
  if(numero % 2 == 0): # Si el número es par, ejecuta el siguiente bloque de código
    print(f"El número {numero} es par")
  else: # Si el número es impar, no debe imprimir nada
    continue  #Pasa al siguiente número

  suma_total += numero

  if(numero == 8):
    break # Sale del bucle cuando llegué al número 8

print("------------------------------------------")
print(f"La sumatoria total es: {suma_total}")
````

````python
estudiantes = ['Yis', 'Cristian', 'Jennifer','Juana']
for estudiante in estudiantes:
  if estudiante == 'Cristian':
    continue # Omite pero no detiene el ciclo
  print(f"Hola {estudiante}")
````

### Bucles While (Iteración basada en una condición booleana)

````python
#El bucle while evalua una condición y ejecuta el bloque de código interno solamente si la condicion es verdadera. Se detendrá cuando la condición cambie a Falso
contador = 0
while contador <= 10:
  print(f"El valor del contador es: {contador}")
  contador += 1
````
